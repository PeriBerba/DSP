```python
import numpy as np
from scipy.fft import fft
import matplotlib.pyplot as plt


# Calculate the Fast Fourier Transform (FFT) of the temperature data
fft_result = fft(temperature_data)

# Get the frequencies corresponding to the FFT result
frequencies = np.fft.fftfreq(len(temperature_data))

# Find the index of the maximum non-zero magnitude in the FFT result
fundamental_frequency_index = np.argmax(np.abs(fft_result[1:len(temperature_data)//2])) + 1

# Calculate the fundamental frequency in Hertz
fundamental_frequency_hz = abs(frequencies[fundamental_frequency_index])

print("Fundamental Frequency:", fundamental_frequency_hz, "Hz")

# Plot the FFT magnitude and highlight the fundamental frequency
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(temperature_data)
plt.title("Temperature Data")
plt.xlabel("Time")
plt.ylabel("Temperature")

plt.subplot(2, 1, 2)
plt.plot(frequencies[1:len(temperature_data)//2], np.abs(fft_result[1:len(temperature_data)//2]))
plt.title("FFT Magnitude")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")

# Highlight the fundamental frequency
plt.axvline(fundamental_frequency_hz, color='r', linestyle='--', label=f'Fundamental Frequency: {fundamental_frequency_hz} Hz')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Create a sample dataset with hourly temperature readings from Oct 20 to Nov 5
start_date = '2023-10-20'
end_date = '2023-11-05'
date_range = pd.date_range(start=start_date, end=end_date, freq='H')
hourly_temperatures = np.random.uniform(25, 35, len(date_range))
temperature_data = pd.Series(hourly_temperatures, index=date_range)
daily_mean_temperatures = temperature_data.resample('D').mean()

plt.figure(figsize=(12, 6))

# Plot hourly temperatures as points
plt.plot(temperature_data.index, temperature_data.values, marker='o', markersize=3, label='Hourly Temperature', linestyle='-', color='blue')

# Plot daily mean temperatures as a line
plt.plot(daily_mean_temperatures.index, daily_mean_temperatures.values, label='Daily Mean Temperature', linestyle='-', color='red')

plt.xlabel('Date')
plt.ylabel('Temperature (Â°F)')
plt.title('Hourly and Daily Mean Temperature')
plt.legend()
plt.grid()
plt.show()
```
